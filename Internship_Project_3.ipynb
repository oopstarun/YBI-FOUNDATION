**Bank Customer Churn Model**

**Objective**

The objective of a Bank Customer Churn Model is to predict which customers are likely to leave the bank in the near future. By accurately identifying these customers, the bank can take proactive measures to retain them, thereby reducing churn rates and increasing customer lifetime value.

**Data Source**

Ybi Foundation Github

**Import Library**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r'https://raw.githubusercontent.com/YBI-Foundation/Dataset/main/Bank%20Churn%20Modelling.csv')
```
